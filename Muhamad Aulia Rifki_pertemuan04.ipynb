```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n-1):
        for j in range(n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

arr = [100, 20, 60, 90, 40, 30, 10]
print("Sebelum diurutkan:", arr)
arr = bubble_sort(arr)
print("Setelah diurutkan:", arr)
```

```python
def insertion_sort(arr):
  for i in range(1, len(arr)):
    key = arr[i]
    j = i - 1
    while j >= 0 and arr[j] > key:
      arr[j + 1] = arr[j]
      j -= 1
    arr[j + 1] = key

data = [89, 12, 57, 16, 25, 11, 75]
print("Data sebelum diurutkan:", data)
insertion_sort(data)
print("Data setelah diurutkan:", data)
```

```python
def selection_sort(arr):
  n = len(arr)
  for i in range(n-1):
    min_idx = i
    for j in range(i+1, n):
      if arr[j] < arr[min_idx]:
        min_idx = j
    arr[i], arr[min_idx] = arr[min_idx], arr[i]

data = [89, 12, 57, 16, 25]
print("Data sebelum diurutkan:", data)

selection_sort(data)
print("Data setelah diurutkan:", data)
```

```python
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1

arr = ['y', 'u', 'i', 'w', 'o', 'a', 'q', 'u', 'j', 'p']
target = 'a'
index = linear_search(arr, target)
if index != -1:
    print("Huruf 'a' ditemukan pada posisi", index)
else:
    print("Huruf 'a' tidak ditemukan")
```

```python
letters = ['y', 'u', 'i', 'w', 'o', 'a', 'q', 'u', 'j', 'p']
letters.sort()
def binary_search(arr, target):
    left = 0
    right = len(arr) - 1

    while left <= right:
        mid = left + (right - left) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] > target:
            right = mid - 1
        else:
            left = mid + 1
    return -1

target = 'a'
index = binary_search(letters, target)
if index != -1:
    print(f"Huruf '{target}' ditemukan pada indeks {index} dalam deret yang diurutkan.")
else:
    print(f"Huruf '{target}' tidak ditemukan dalam deret.")
```

```python
letters = ['y', 'u', 'i', 'w', 'o', 'a', 'q', 'u', 'j', 'p']

letters.sort()
def interpolation_search(arr, target):
    low = 0
    high = len(arr) - 1

    while low <= high and target >= arr[low] and target <= arr[high]:
        pos = low + ((high - low) // (ord(arr[high]) - ord(arr[low])) * (ord(target) - ord(arr[low])))
        if arr[pos] == target:
            return pos
        elif arr[pos] < target:
            low = pos + 1
        else:
            high = pos - 1
    return -1

target = 'u'
index = interpolation_search(letters, target)
if index != -1:
    print(f"Huruf '{target}' ditemukan pada indeks {index} dalam deret yang diurutkan.")
else:
    print(f"Huruf '{target}' tidak ditemukan dalam deret.")
```
